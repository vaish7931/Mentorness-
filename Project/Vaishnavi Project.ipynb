```python
#2.Crop-specific Analysis
```

```python
rice_data = df[df['RICE AREA (1000 ha)'] == 'Rice']
wheat_data = df[df['RICE AREA (1000 ha)'] == 'Wheat']
pulses_data = df[df['RICE AREA (1000 ha)'].isin(['Pulses', 'Lentils', 'Chickpeas'])]
```

```python
rice_production_trend = rice_data.groupby('Year')['RICE PRODUCTION (1000 tons)'].sum()
```

```python
rice_production_trend
```

```python
district_disparities = df.groupby('Dist Name')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)
print(district_disparities.head())
```

```python
plt.bar(district_disparities.head().index, district_disparities.head().values)
plt.xlabel('Dist Name')
plt.ylabel('Total Production')
plt.title('Top 5 Districts by Production')
plt.xticks(rotation=45)
plt.show()
```

```python
kharif_data = df[df['KHARIF SORGHUM AREA (1000 ha)'] == 'Kharif']
rabi_data = df[df['RABI SORGHUM AREA (1000 ha)'] == 'Rabi']
```

```python
kharif_data
```

```python
rabi_data
```

```python
kharif_production = kharif_data.groupby('RICE AREA (1000 ha)')['RICE PRODUCTION (1000 tons)'].sum()
rabi_production = rabi_data.groupby('RICE AREA (1000 ha)')['RICE PRODUCTION (1000 tons)'].sum()
```

```python
kharif_production
```

```python
rabi_production
```

```python
plt.plot(kharif_production, label='Kharif')
plt.plot(rabi_production, label='Rabi')
plt.xlabel('Crop')
plt.ylabel('Production')
plt.title('Seasonal Crop Production')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
